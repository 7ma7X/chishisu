```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(arr):
    return (10 * arr[0] * arr[0] + arr[1] * arr[1])
```

```python
# gradient of f
def g(arr):
    return np.array([20 * arr[0], 2 * arr[1]])
```

```python
alpha = 0.5
beta  = 0.8
```

```python
eps = []
x   = []

# initial value
x.append(np.array([1, 5]))

i = 0

while True:
    if i == 10:
        break
    
    eps.append(1)
    
    while f(x[i] - eps[i] * g(x[i])) - f(x[i]) > - alpha * eps[i] * np.dot(g(x[i]), g(x[i])):
        eps[i] *= beta
    
    x.append(x[i] - eps[i] * g(x[i]))
    print("i={} epsilon={} x={}".format(i, eps[i], x[i]))
    i += 1
```

```python
# preparations for plotting

X = []
Y = []
for (p, q) in x:
    X.append(p)
    Y.append(q)
    
n = 500
s = np.linspace(-5, 5, n)
t = np.linspace(-5, 5, n)
S, T = np.meshgrid(s, t)
U = f(np.array([S, T]))
```

```python
plt.figure(figsize=(6, 6))
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.gca().set_aspect('equal')
plt.contour(S, T, U)
plt.plot(X, Y, 'r')
```
