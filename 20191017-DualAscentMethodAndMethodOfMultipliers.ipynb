```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(arr):
    return 3 * arr[0] * arr[0] + 2 * arr[1] * arr[1]
```

```python
def g(arr):
    return arr[0] + arr[1] - 1
```

```python
def omega(la):
    return ((-5) * la * la) / 24 - la
```

```python
# Dual Ascent Method

x = []
la = [] # lambda
    
eps = []
alpha = 0.5
beta  = 0.8

# initialize
x.append(np.array([0.0, 0.0]))
la.append(0.0)

i = 0

while True:
    if i == 10:
        break
        
    # minimize Lagrangian
    x.append(np.array([(-la[i]) / 6.0, (-la[i]) / 4.0]))
    
    # decide epsilon (step size) by backtracking line search
    eps.append(1.0)
    
    while (omega(la[i] + eps[i] * g(x[i+1])) - omega(la[i])) < alpha * eps[i] * g(x[i+1]) * g(x[i+1]):
        eps[i] *= beta
    
    la.append(la[i] + eps[i] * g(x[i+1]))
    print("i={} epsilon={} lambda={} x={}".format(i, eps[i], la[i], x[i]))
    i += 1
```

```python
# Method of Multipliers (c=5)

x2 = []
la2 = [] # lambda
    
c = 5.0

# initialize
x2.append(np.array([0.0, 0.0]))
la2.append(0.0)

i = 0

while True:
    if i == 10:
        break
        
    # minimize Lagrangian
    x2.append(np.array([(c - la2[i]) / (6.0 + (5.0 * c / 2.0)), (c - la2[i]) / (4.0 + (5.0 * c / 3.0))]))
    
    la2.append(la2[i] + c * g(x2[i+1]))
    print("i={} lambda={} x={}".format(i, la2[i], x2[i]))
    i += 1
```

```python
# Method of Multipliers (c=1)

x3 = []
la3 = [] # lambda
    
c = 1.0

# initialize
x3.append(np.array([0.0, 0.0]))
la3.append(0.0)

i = 0

while True:
    if i == 10:
        break
        
    # minimize Lagrangian
    x3.append(np.array([(c - la3[i]) / (6.0 + (5.0 * c / 2.0)), (c - la3[i]) / (4.0 + (5.0 * c / 3.0))]))
    
    la3.append(la3[i] + c * g(x3[i+1]))
    print("i={} lambda={} x={}".format(i, la3[i], x3[i]))
    i += 1
```

```python
# Method of Multipliers (c=10)

x4 = []
la4 = [] # lambda
    
c = 10.0

# initialize
x4.append(np.array([0.0, 0.0]))
la4.append(0.0)

i = 0

while True:
    if i == 10:
        break
        
    # minimize Lagrangian
    x4.append(np.array([(c - la4[i]) / (6.0 + (5.0 * c / 2.0)), (c - la4[i]) / (4.0 + (5.0 * c / 3.0))]))
    
    la4.append(la4[i] + c * g(x4[i+1]))
    print("i={} lambda={} x={}".format(i, la4[i], x4[i]))
    i += 1
```

```python
plt.figure(figsize=(6, 6))
plt.ylim([-2.5, 0])
iter_num = [i for i in range(10)]
plt.plot(iter_num, la[:10], label="Dual Ascent Method")
plt.plot(iter_num, la2[:10], label="Method of Multipliers (c=5)")
plt.plot(iter_num, la3[:10], label="Method of Multipliers (c=1)")
plt.plot(iter_num, la4[:10], label="Method of Multipliers (c=10)")
plt.legend()
```

```python
# The value lambda converged more quickly in Method of Multipliers (c=5, 10) than Dual Ascent Method, but less quickly in Method of Multipliers (c=1)
```
