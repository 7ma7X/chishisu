```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
# create chi-square specimen

chi = np.random.chisquare(2, 10000)
chi_ave = [sum(chi[:i]) / i for i in range(1, 10001)]
```

```python
plt.figure(figsize=(8, 6))
n = [i for i in range(1, 10001)]
ave = [2 for i in range(1, 10001)]

plt.ylim([0, 4])
plt.plot(n, chi_ave, label="actual")
plt.plot(n, ave, color = "black", linestyle = ":", label="Î¼(2.0)")
plt.legend()
```

```python
chi_x_1 = chi[:1000]
chi_x_2 = [sum(chi[2*i:2*(i+1)]) / 2 for i in range(1000)]
chi_x_5 = [sum(chi[5*i:5*(i+1)]) / 5 for i in range(1000)]
chi_x_10 = [sum(chi[10*i:10*(i+1)]) / 10 for i in range(1000)]
```

```python
chi_z_1 = [(chi_x_1[i] - 2) / (4 / sqrt(1)) for i in range(1000)]
chi_z_2 = [(chi_x_2[i] - 2) / (4 / sqrt(2)) for i in range(1000)]
chi_z_5 = [(chi_x_5[i] - 2) / (4 / sqrt(5)) for i in range(1000)]
chi_z_10 = [(chi_x_10[i] - 2) / (4 / sqrt(10)) for i in range(1000)]
```

```python
plt.figure(figsize=(8, 6))
plt.hist(chi_z_1, bins=40, label="n = 1")
plt.legend()
1
```

```python
plt.figure(figsize=(8, 6))
plt.hist(chi_z_2, bins=40, label="n = 2")
plt.legend()
2
```

```python
plt.figure(figsize=(8, 6))
plt.hist(chi_z_5, bins=40, label="n = 5")
plt.legend()
5
```

```python
plt.figure(figsize=(8, 6))
plt.hist(chi_z_10, bins=40, label="n = 10")
plt.legend()
10
```

```python
# t distribution

t = np.random.standard_t(2, 10000)
t_ave = [sum(t[:i]) / i for i in range(1, 10001)]
```

```python
plt.figure(figsize=(8, 6))
n = [i for i in range(1, 10001)]
ave = [0 for i in range(1, 10001)]

plt.ylim([-2, 2])
plt.plot(n, t_ave, label="actual")
plt.plot(n, ave, color = "black", linestyle = ":", label="Î¼(0.0)")
plt.legend()
```

```python
t_x_1 = t[:1000]
t_x_2 = [sum(t[2*i:2*(i+1)]) / 2 for i in range(1000)]
t_x_5 = [sum(t[5*i:5*(i+1)]) / 5 for i in range(1000)]
t_x_10 = [sum(t[10*i:10*(i+1)]) / 10 for i in range(1000)]
```

```python
t_z_1 = [t_x_1[i] / (1 / sqrt(1)) for i in range(1000)]
t_z_2 = [t_x_2[i] / (1 / sqrt(2)) for i in range(1000)]
t_z_5 = [t_x_5[i] / (1 / sqrt(5)) for i in range(1000)]
t_z_10 = [t_x_10[i] / (1 / sqrt(10)) for i in range(1000)]
```

```python
plt.figure(figsize=(8, 6))
plt.hist(t_z_1, bins=40, label="n = 1")
plt.legend()
1
```

```python
plt.figure(figsize=(8, 6))
plt.hist(t_z_2, bins=40, label="n = 2")
plt.legend()
2
```

```python
plt.figure(figsize=(8, 6))
plt.hist(t_z_5, bins=40, label="n = 5")
plt.legend()
5
```

```python
plt.figure(figsize=(8, 6))
plt.hist(t_z_10, bins=40, label="n = 10")
plt.legend()
10
```
