```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = np.random.normal(size = 1000000)
data2 = np.random.normal(size = 1000000)
data3 = np.random.normal(size = 1000000)
data4 = np.random.normal(size = 1000000)
data5 = np.random.normal(size = 1000000)
```

```python
# degree of freedom = 1

y_1 = [x1 * x1 for x1 in data1]
plt.hist(y_1, bins=100)
1
```

```python
# degree of freedom = 2

y_2 = [x1 * x1 + x2 * x2 for x1, x2 in zip(data1, data2)]
plt.hist(y_2, bins=100)
2
```

```python
# degree of freedom = 3

y_3 = [x1 * x1 + x2 * x2 + x3 * x3 for x1, x2, x3 in zip(data1, data2, data3)]
plt.hist(y_3, bins=100)
3
```

```python
# degree of freedom = 4

y_4 = [x1 * x1 + x2 * x2 + x3 * x3 + x4 * x4 for x1, x2, x3, x4 in zip(data1, data2, data3, data4)]
plt.hist(y_4, bins=50)
4
```

```python
# degree of freedom = 5

y_5 = [x1 * x1 + x2 * x2 + x3 * x3 + x4 * x4 + x5 * x5 for x1, x2, x3, x4, x5 in zip(data1, data2, data3, data4, data5)]
plt.hist(y_5, bins=50)
5
```
